```python
# Matplotlib vs seaborn
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
# Create some data
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)

# plot the data with matplotlib defaults
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left')
```

```python
import seaborn as sns
sns.set()
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left')
```

```python
# Classic matplotlib histogram

data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
data
```

```python
for col in 'xy':
    plt.hist(data[col],density=True,alpha=0.5)
```

```python
# Rather than a histogram we can get a smooth estimate of the distribution useing a kernel density estimation
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
# combine both with distplot
sns.distplot(data['x'])
sns.distplot(data['y'])
```

```python
# If we pass a full 2d dataset to kdeplot we will get a 2d visualisation
sns.kdeplot(data=data, x='x', y='y')
```

```python
with sns.axes_style('white'):
    sns.jointplot('x', 'y', data=data, kind='kde')
```

```python
# Pair plots
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue='species', size=2.5)
```

```python
# Box plot
sns.axes_style(style='ticks')
g = sns.factorplot("species", "sepal_length", data=iris, kind="box")
```

```python
# Regression
sns.jointplot("petal_length", "petal_width", data=iris, kind='reg')
```
